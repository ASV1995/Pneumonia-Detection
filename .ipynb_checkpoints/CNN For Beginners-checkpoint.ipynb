```python
import tensorflow as tf
```

```python
GPUS = tf.config.experimental.list_physical_devices('GPU')
for GPU in GPUS:
    tf.config.experimental.set_memory_growth(GPU,True)
    
# This is to avoid any OOM (Out of Memory) issues by setting GPI Memory Consumption.
```

```python
import os
# Helps interact with files from the user's operating systems.
import cv2
import numpy as np
# Essential for normalical computing such as matrices which are commonly used in machine learning.
import matplotlib.pyplot as plt
```

```python
dataset_path = 'C:\\Users\\15149\\Desktop\\Chest Pneumonia Image Classification\\Data'

train_dir = os.path.join(dataset_path, 'train')
test_dir = os.path.join(dataset_path, 'test')
val_dir = os.path.join(dataset_path, 'val')
```

```python
from keras.preprocessing.image import ImageDataGenerator # Used for image data augmentation during training
```

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,  # Normalizing pixel values to a consistent and standardized range, such as between 0 and 1
    rotation_range=20,  # Rotate images up to 20 degrees.
    width_shift_range=0.2,  # Shift image width.
    height_shift_range=0.2,  # Shift image height.
    shear_range=0.2,  # Apply shear transformation (distorts the shape of an image)
    zoom_range=0.2,  # Zoom in/out.
    horizontal_flip=True,  # Flip horizontally.
    fill_mode='nearest'  # How to fill newly created pixels when image is transformed
)
```

```python
test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
val_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
```

```python
train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(256, 256),
    batch_size=20, # Parameter for setting number of samples to be processed in a single iteration during training of a neural network
    class_mode='binary'
    class_mode='binary',
)

valid_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=(256, 256),
    batch_size=20,
    class_mode='binary'
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(256, 256),
    batch_size=20,
    class_mode='binary'
)
```

```python
# Training set: Must use for training your model.
# Validation set: Must use for tuning parameters and evaluating model performance during training.
# Test set: Not seen by the model during training, used to TEST the model's performance.
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(256, 256, 3)), #Process 2D input for feature extraction in images
    tf.keras.layers.MaxPooling2D(2, 2),
    
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    
    tf.keras.layers.Conv2D(128, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    
    tf.keras.layers.Flatten(), #Reshape input into a 1D array from Convolutional layers to dense layers.
    
    tf.keras.layers.Dense(128, activation='relu'),# Dense layers connects all neurons 
    tf.keras.layers.Dense(1, activation='sigmoid')  # Binary classification, so using sigmoid activation!
    
    # Connects all neurons
])
```

```python
!pip install --upgrade tensorflow
# WARNING:tensorflow:From C:\Users\15149\anaconda3\Lib\site-packages\keras\src\optimizers\__init__.py:309: The name tf.train.Optimizer is deprecated. Please use tf.compat.v1.train.Optimizer instead.
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
hist = model.fit(train_generator, epochs=20, validation_data = valid_generator)
```

```python
fig = plt.figure()
plt.plot(hist.history['loss'], color = 'green', label = 'loss')
plt.plot(hist.history['val_loss'], color = 'blue', label = 'val_loss')
fig.suptitle('Loss', fontsize = 20)
plt.legend(loc='upper left')
plt.show()
```

```python
fig = plt.figure()
plt.plot(hist.history['accuracy'], color = 'green', label = 'accuracy')
plt.plot(hist.history['val_accuracy'], color = 'blue', label = 'val_accuracy')
fig.suptitle('Accuracy', fontsize = 20)
plt.legend(loc='upper left')
plt.show()
```

```python
from tensorflow.keras.metrics import Precision, BinaryAccuracy
```

```python
binary_accuracy, precision = model.evaluate(test_generator)

print(f'Binary Accuracy: {binary_accuracy}') # Evaluate the performance of the model to Number of Correct Predictions/Total Predictions
print(f'Precision: {precision}') # Measures accuracy of positive predictions with respect to TP (True Positive) + FP (False Positive)
```

```python
resize = tf.image.resize(img, (256, 256))
plt.imshow(resize.numpy().astype(int))
plt.show()
```
